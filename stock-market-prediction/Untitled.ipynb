```python
import nltk
import jieba
import heapq
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer


f = open('news.txt', mode='r', encoding='utf-8')
news = [eval(i) for i in f.readlines()]
f.close()

f = open('train.txt', mode='r', encoding='utf-8')
train = [i.split() for i in f.readlines()]
f.close

f = open('test.txt', mode='r', encoding='utf-8')
test = [i.split() for i in f.readlines()]
f.close

f = open('stopwords.txt', mode='r', encoding='utf-8')
stopWords = [i for i in f.read() if i != '\n']
f.close()
```

```python

def content(test, news):
    test_copy = list(test)
    for i in range(len(test_copy)):
        id_str = test_copy[i][1].split(',')
        id_list = [int(i) for i in id_str]
        title_list = []
        content_list = []
        for j in range(len(news)):
            a = news[j]
            content = a['content']
            title = a['title']
            id = a['id']
            if id in id_list:
                title_list.append(title)
                content_list.append(content)
        test_copy[i][1] = content_list
        # test_copy[i].append(title_list)
        # test_copy[i].append(content_list)
    return test_copy

new_train = content(train, news)
new_test = content(test, news)
print(type(new_test),len(new_test),
    type(new_test[0]),len(new_test[0]))
```

```python

def del_stopword(train):
    train_copy = list(train)
    for each in train_copy:
        result = ''
        for sentence in each[1]:
            #print(sentence)
            cut = jieba.cut(sentence)
            cut = ' '.join(cut)
            #print(cut)
            cut = cut.split()
            #print(cut)
            temp = ''
            for word in cut:
                if word not in stopWords:
                    temp = temp + word + ' '
            result = result + temp
        each[1] = result
        #print(each[1])
        #break
    return train_copy

del_new_train = del_stopword(new_train)
```

```python
print(del_new_train[0])
```

```python

def txt2str(train):
    str1, str2 = '', ''
    for each in train:
        if each[0] == '+1':
            str1 = str1 + each[1]
        elif each[0] == '-1':
            str2 = str2 + each[1]
    return [str1, str2]

[str1, str2] = txt2str(del_new_train)
```

```python

def tfidf(str1, str2):
    X_test = [str1, str2]
    tfidf = TfidfVectorizer()
    weight = tfidf.fit_transform(X_test).toarray()
    word = tfidf.get_feature_names()
    vocab = []
    #print(word)
    for i in range(len(weight)):
    # 打印每类文本的tf-idf词语权重，第一个for遍历所有文本，
    # 第二个for便利某一类文本下的词语权重
        nums = list(weight[i])
        #print(nums[0:500])
        #print(heapq.nlargest(50, nums))
        idx = list(map(nums.index, heapq.nlargest(500, nums)))
        #print(idx)
        # print(u"-------这里输出第", i, u"类文本的词语tf-idf权重------")
        for j in idx:
            vocab.append(word[j]) 
            # print(word[j])
            # print(word[j], weight[i][j]) # 第i个文本中，第j个次的tfidf值
    return vocab
vocab = tfidf(str1, str2)
# print(vocab)
```

```python
vocab[1]
```

```python

def prepare(train):
    train_copy = list(train)
    for each in train_copy:
        temp = []
        words = each[1].split()
        for word in words:
            if word in vocab:
                temp.append(word)
        each[1] = each[0]
        each[0] = dict([(word, False) for word in temp])
    return train_copy

prepare_del_new_train = prepare(del_new_train)
```

```python
new_train[1]
```

```python

classifier = nltk.NaiveBayesClassifier.train(prepare_del_new_train) #生成分类器
```

```python
test = prepare(del_stopword(new_test))
print(test[0])
```

```python


f = open('result.txt', 'w')
for i in range(len(test)):
    tag = classifier.classify(test[i][0]) # 分类
    f.write(tag+' '+test[i][1]+'\n')
f.close()
```
